```python
import json
#Parsing the provided course information and converting it to JSON format with only course number, name, and description

def extract_course_info(course_text):
    # Extract course number, name, and description from the text
    parts = course_text.split(", Description: ")
    course_title_parts = parts[0].split(" ")
    course_number = course_title_parts[1]
    course_name = " ".join(course_title_parts[2:])
    course_description = parts[1].split(". Prerequisite")[0].strip()
    return {
        "Course Number": course_number,
        "Course Name": course_name,
        "Description": course_description
    }
```
